```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/Lena_lowcontrast.jpg', 0)

if img is None:
    print("Error: Could not load image. Please check the file path.")
else:
    rmin, rmax = np.min(img), np.max(img)

    stretched = ((img - rmin) * 255 / (rmax - rmin)).astype(np.uint8)

    plt.subplot(1, 2, 1); plt.imshow(img, cmap='gray'); plt.title('Original')
    plt.subplot(1, 2, 2);
    plt.imshow(stretched, cmap='gray'); plt.title('Contrast Stretched')
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/Lena_lowcontrast.jpg', 0)

equalized = cv2.equalizeHist(img)
plt.subplot(2,2,1), plt.imshow(img, cmap='gray'),plt.title('Original')
plt.subplot(2,2,2), plt.hist(img.ravel(),
256), plt.title('Original Histogram')

plt.subplot(2,2,3), plt.imshow(equalized,cmap='gray'), plt.title('Equalized')
plt.subplot(2,2,4),
plt.hist(equalized.ravel(), 256), plt.title('Equalized Histogram')
plt.tight_layout()
plt.show()
```
